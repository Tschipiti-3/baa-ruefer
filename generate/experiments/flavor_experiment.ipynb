# 1. LLaMA

```python
import openai
import json
import warnings
warnings.filterwarnings("ignore")

def generate_flavor_output(input):
    ## point openai library to internal endpoint
    openai.api_key = "1234"
    openai.api_base = "http://10.180.132.23:8180/v1"

    messages = [
        {   
            "role": "system", 
            "content": "You get a text from a sports journalist for live commentary. Write the text with the provided flavor."
        },{
            "role": "user",
            "content": f"{input}"
        },
    ]

    model_name = "TheBloke/Llama-2-13B-chat-AWQ"
    response = openai.ChatCompletion.create(
        model=model_name,
        messages=messages,
        temperature=0,
    )
    return response["choices"][0]["message"]["content"]

team_id = 489
flavors = ['humorous', 'sarcastic', 'serious', 'neutral', 'enthusiastic', 'analytical', 'emotional', 'educational', 'upbeat and energetic', 'wordplay', 'bored', 'overacting']

with open('output_2.json') as json_file:
    input_json = json.load(json_file)
```

# 2. GPT-4 

```python
from openai import OpenAI
import json
import warnings
warnings.filterwarnings("ignore")

def generate_gpt_flavor(input):
        client = OpenAI(
            # defaults to os.environ.get("OPENAI_API_KEY")
            api_key="API_KEY",
        )

        response = client.chat.completions.create(
            messages=[
                {"role": "system", "content": "You get a text from a sports journalist for live commentary. Write the text with the provided flavor."},
                {"role": "user", "content": f"{input}"},
            ],
            model="gpt-4",
            temperature=0.9,
        )
        return response.choices[0].message.content

team_id = 489
flavors = ['humorous', 'sarcastic', 'serious', 'neutral', 'enthusiastic', 'analytical', 'emotional', 'educational', 'upbeat and energetic', 'wordplay', 'bored', 'overacting']

with open('output_2.json') as json_file:
    input_json = json.load(json_file)
```
